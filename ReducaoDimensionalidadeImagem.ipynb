Converte uma imagem para escala de cinza

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
img_file = files.upload()
```

```python
image = plt.imread("image.jpg")
plt.imshow(image)
plt.plot()
```

```python
def grayscale(img):
  r_channel = img[:,:,0]
  g_channel = img[:,:,1]
  b_channel = img[:,:,2]

  r_weight = 0.299
  g_weight = 0.587
  b_weight = 0.114

  grayscale_img = r_channel * r_weight + g_channel * g_weight + b_channel * b_weight
  return grayscale_img
```

```python
plt.imshow(grayscale(image), cmap="gray")
plt.plot()
plt.imsave("gray.jpg", grayscale(image), cmap="gray")
```
